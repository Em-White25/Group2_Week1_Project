```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_titanic = pd.read_csv('../data/processed/Cleaned_Titanic_Dataset.csv', index_col= "S/N")
df_titanic.head()
```

```python
sns.barplot(x=df_titanic['pclass'], y=df_titanic['survived'], data=df_titanic)
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')
plt.xlabel('Passenger Class')
plt.xticks([0, 1, 2], ['1st', '2nd', '3rd'])
plt.show()
```

```python
sns.barplot(x=df_titanic['sex'], y=df_titanic['survived'], data=df_titanic)
plt.title('Survival Rate by Sex')
plt.ylabel('Survival Rate')
plt.xlabel('Sex')
plt.show()
```

```python
sns.barplot(x=df_titanic['embarked'], y=df_titanic['survived'], data=df_titanic)
plt.title('Survival Rate by Embarked Port')
plt.ylabel('Survival Rate')
plt.xlabel('Embarked Port')
plt.show()
```

```python
sns.histplot(data=df_titanic, x=df_titanic['age'], hue=df_titanic['survived'], kde=True)
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
sns.histplot(data=df_titanic, x=df_titanic['fare'], hue=df_titanic['survived'], kde=True)
plt.title('Fare Distribution by Survival')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.show()
```

```python
sns.boxplot(x=df_titanic['survived'], y=df_titanic['fare'], data=df_titanic)
plt.title('Fare Distribution by Survival')
plt.xlabel('Survived (0=No, 1=Yes)')
plt.ylabel('Fare')
plt.show()
```

```python
sns.catplot(x=df_titanic['pclass'], y=df_titanic['survived'], hue=df_titanic['sex'], data=df_titanic, kind='bar')
plt.title('Survival Rate by Passenger Class and Sex')
plt.ylabel('Survival Rate')
plt.xlabel('Passenger Class')
plt.xticks([0, 1, 2], ['1st', '2nd', '3rd'])
plt.show()
```

```python
sns.catplot(x=df_titanic['pclass'], y=df_titanic['survived'], hue=df_titanic['embarked'], data=df_titanic, kind='bar')
plt.title('Survival Rate by Passenger Class and Embarked Port')
plt.ylabel('Survival Rate')
plt.xlabel('Passenger Class')
plt.xticks([0, 1, 2], ['1st', '2nd', '3rd'])
plt.show()
```

```python
sns.catplot(x=df_titanic['sex'], y=df_titanic['survived'], hue=df_titanic['embarked'], data=df_titanic, kind='bar')
plt.title('Survival Rate by Sex and Embarked Port')
plt.ylabel('Survival Rate')
plt.xlabel('Sex')
plt.show()
```

```python
sns.boxplot(x=df_titanic['pclass'], y=df_titanic['age'], hue=df_titanic['survived'], data=df_titanic)
plt.title('Age Distribution by Passenger Class and Survival')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.xticks([0, 1, 2], ['1st', '2nd', '3rd'])
plt.show()
```

```python
sns.boxplot(x=df_titanic['pclass'], y=df_titanic['fare'], hue=df_titanic['survived'], data=df_titanic)
plt.title('Fare Distribution by Passenger Class and Survival')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.xticks([0, 1, 2], ['1st', '2nd', '3rd'])
plt.show()
```

```python
sns.scatterplot(x=df_titanic['age'], y=df_titanic['fare'], hue=df_titanic['survived'], data=df_titanic)
plt.title('Age vs. Fare, Colored by Survival')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
sns.countplot(x=df_titanic['pclass'], data=df_titanic)
plt.title('Passenger Count by Class')
plt.xlabel('Passenger Class')
plt.xticks([0, 1, 2], ['1st', '2nd', '3rd'])
plt.show()
```

```python
sns.countplot(x=df_titanic['sex'], data=df_titanic)
plt.title('Passenger Count by Sex')
plt.xlabel('Sex')
plt.show()
```

```python
sns.countplot(x=df_titanic['embarked'], data=df_titanic)
plt.title('Passenger Count by Embarked Port')
plt.xlabel('Embarked Port')
plt.show()
```
