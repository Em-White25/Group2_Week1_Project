```python
import pandas as pd
import numpy as np
```

```python
df_titanic = pd.read_csv('../data/raw/Titanic dataset.csv')

df_titanic.head(5)
```

```python
#created a copy of the original dataset
df_titanic_copy = df_titanic
df_titanic_copy
```

```python
df_titanic_copy.index.rename("S/N", inplace = True)
```

```python
#drop name, home.dest, cabin, ticket, sibsp, parch, boat, body columns
df_titanic_copy.drop(columns=['name', 'home.dest', 'cabin', 'ticket', "sibsp", "parch", "boat", "body"], inplace=True, errors='ignore')
df_titanic_copy.head()


```

```python
#remove duplicate values
df_titanic_copy.drop_duplicates(inplace=True)
```

```python
df_titanic_copy.isnull().sum()
```

```python
# Compute mean age for each passenger class
pclass_means_age = df_titanic_copy.groupby('pclass')['age'].mean().round().astype(int)


def meanAge(cols, pclass_means_age):
    age, pclass = cols
    
    if pd.isnull(age):
        return int(round(pclass_means_age[pclass]))  # Round and convert to integer
    elif age < 1:
        return 1
    else:
        return int(round(age))  # Ensure age is also a whole number
    
df_titanic_copy["age"] = df_titanic_copy[["age", "pclass"]].apply(meanAge, axis =1, pclass_means_age = pclass_means_age)

```

```python
pclass_means_fare = df_titanic_copy.groupby('pclass')['fare'].mean().round().astype(int)


def meanAge(cols, pclass_means_fare):
    age, pclass = cols
    
    if pd.isnull(age):
        return int(round(pclass_means_fare[pclass]))  # Round and convert to integer
    else:
        return int(round(age))  # Ensure fare is also a whole number
    
df_titanic_copy["fare"] = df_titanic_copy[["fare", "pclass"]].apply(meanAge, axis =1, pclass_means_fare = pclass_means_fare)
```

```python
df_titanic_copy.dropna(inplace=True)
df_titanic_copy.isnull().sum()
```
